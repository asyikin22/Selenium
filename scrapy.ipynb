```python
import pandas as pd 

df = pd.read_csv(r"Goodreads_PH.csv")
df.columns = df.columns.str.strip()
```

```python
df.head()
```

```python
#create scrapy spider in notebook 

import scrapy
from scrapy.crawler import CrawlerProcess
from scrapy.utils.project import get_project_settings

#Define the spider 
class GoodreadSpider(scrapy.Spider):
    name = 'goodreads'
    
    def start_requests(self):
        for index, row in df.iterrows():
            url = row['URL']
            print(f'Index: {index}, URL:{url}')
            yield scrapy.Request(url=row['URL'], callback=self.parse)
    
    def parse(self, response):
        title = response.css('h1#bookTitle::text').get().strip()
        rating = response.css('span[itemprop="ratingValue"]::text').get().strip()
        yield {
            'ID': response.url.split('/')[-1],
            'Title': title, 
            'Rating': rating, 
            'URL': response.url
        }

# configure and run the crawler 
settings = {
    'FEEDS': {
        'books_ratings.csv': {
            'format':'csv',
            'overwrite': True
        },
    },
    'DOWNLOAD_DELAY': 1, 
    'USER_AGENT': 'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/128.0.0.0 Mobile Safari/537.36'
}

process = CrawlerProcess(settings)
process.crawl(GoodreadSpider)
process.start()

#load the scraped data
scraped_df = pd.read_csv('books_ratings.csv')
scraped_df.head()
```
